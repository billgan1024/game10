```sage
%display latex
fov, aspect, near = var('fov aspect near')

x1, y1, z1, x2, y2, z2, x3, y3, z3 = var('x1 y1 z1 x2 y2 z2 x3 y3 z3')
p1 = vector([x1, y1, z1])
p2 = vector([x2, y2, z2])
p3 = vector([x3, y3, z3])
a, b, c = var('a b c')
p = a*p1 + b*p2 + c*p3

f1, f2, f3 = var('f1 f2 f3')

def ndc(v):
    return vector([v[0]/(v[2]*tan(fov/2)), aspect * v[1]/(v[2]*tan(fov/2)), near/v[2]])

def bary(x, y, x1, y1, x2, y2, x3, y3):
    T = matrix([[x1-x3, x2-x3],
                [y1-y3, y2-y3]])
    t = T.inverse() * vector([x - x3, y - y3])
    return t[0], t[1], 1 - t[0] - t[1]

alpha, beta, gamma = bary(ndc(p)[0], ndc(p)[1], 
     ndc(p1)[0], ndc(p1)[1], 
     ndc(p2)[0], ndc(p2)[1], 
     ndc(p3)[0], ndc(p3)[1])


assume(a+b+c == 1)

# what is the interpolated value of p? it's a*f1 + b*f2 + c*f3, which is also equal to 
# the perspective correct interpolation formula:


# clip w is just the z value (which is > 0 via clipping)
bool(a*f1 + b*f2 + c*f3 == (alpha*f1/z1 + beta*f2/z2 + gamma*f3/z3)/(alpha/z1 + beta/z2 + gamma/z3))
```

```sage
# also check depth linear interpolation is okay
bool(ndc(p)[2] == alpha*ndc(p1)[2] + beta*ndc(p2)[2] + gamma*ndc(p3)[2])
```
